## Question number 7
- Select all the "Lotus" cars from the dataset.
- Select only columns "Engine HP", "Engine Cylinders".
- Now drop all duplicated rows using drop_duplicates method (you should get a dataframe with 9 rows).
- Get the underlying NumPy array. Let's call it X
- Compute matrix-matrix multiplication between the transpose of $X$ and $X$. To get the transpose, use X.T. Let's call the result $XTX$.
- Invert XTX.
- Create an array y with values [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800].
- Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
- What's the value of the first element of w?

```python
df[df['Make']=='Lotus']  # selects all 'Lotus' cars from the dataset
```

```python
df1= df[df['Make']=='Lotus'][['Engine HP','Engine Cylinders']] #selects only columns "Engine HP", "Engine Cylinder
df1
```

```python
df2=df1.drop_duplicates('Engine HP')
df2
```

```python
x = df2.to_numpy() #dataframe is converted into an array
x
```

```python
# transpose of x
xT=df2.T.to_numpy()
xT
```

```python
XTX =xT.dot(x)
XTX
```

```python
np.linalg.det(XTX)  # just to check that the matrix is invertible
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
XTX.dot(np.linalg.inv(XTX))  #just to check that we get an identity matrix from this product
```

```python
# Create an array y with values [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800].
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y
```

```python
# Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
A = XTX_inv.dot(xT)
A
```

```python
w = A.dot(y)
w
```

## Answer to question number 7:
The value of first element of w is 4.5949
